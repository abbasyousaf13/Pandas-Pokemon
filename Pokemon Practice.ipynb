# Feature Engineering
### Add Column

```python
attack_mean = df["Attack"].mean()

def set_attack(val):
    if val < attack_mean:
        return "Attack Low"
    elif val == attack_mean:
        return "Attack Neutral"
    else:
        return "Attack High"
```

```python
df["Attack_high_low"] = df["Attack"].apply(set_attack)
df
```
